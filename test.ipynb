```python
#tf_idf["similarity"] = ""
# for i in range(len(tf_idf)):
#     try:
#         tf_idf["similarity"][i] = model.wv.similarity(w1=tf_idf.word[i],w2="悲しい")
#     except:
#         tf_idf["similarity"][i] = float("inf")
#         continue
import pandas as pd
tf_idf  =pd.read_csv("tf_idf.csv",encoding="shift-jis")
sim_arr = []
for word in tf_idf.word:
    try:
        sim_arr.append(model.wv.similarity(w1=word,w2="愛してる"))
    except:
        sim_arr.append((-1)*float("inf"))

tf_idf["similarity"] = sim_arr
```

```python
tf_idf = tf_idf.sort_values("TF-IDF",ascending=False)
print(tf_idf)
```

```python
import sklearn.preprocessing as sp

# mm.fit_transform(l)
# ValueError: Expected 2D array, got 1D array instead:
tf_idf["TF-IDF"] = sp.minmax_scale(tf_idf["TF-IDF"])

#tf_idf["TF-IDF"] = stats.zscore(tf_idf["TF-IDF"])
tf_idf["emotion"] = tf_idf["TF-IDF"]*tf_idf["similarity"]
tf_idf = tf_idf.sort_values("emotion",ascending=False)
```
